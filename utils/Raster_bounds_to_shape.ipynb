```python
from pytorch_segmentation.utils.helper import raster_bounds_to_shape
```

```python
import sys 
import os
from pathlib import Path
```

```python
path = "/home/jovyan/work/satellite_data/tmp/"
areas = range(22,35)
years = range(2008,2019)
```

```python
for a in areas:
    for y in years:
        fname = os.path.join(path,str(a),str(y)+"_cog.tif")
        spath = os.path.join(path,"shapes",str(a))
        if os.path.isfile(fname):
            print(f"{str(a)} {str(y)}")
            Path(spath).mkdir(parents=True, exist_ok=True)
            sname  = os.path.join(spath,str(y)+".shp")
            raster_bounds_to_shape(fname,sname)
```
