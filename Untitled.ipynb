```python
class RasterD(RasterDataset):
    filename_glob = "30_cog_2020.tif" #"*.tif" 
    
    #filename_regex = r"^.{6}_(?P<date>\d{8}T\d{6})_(?P<band>B0[\d])"
    #date_format = "%Y%m%dT%H%M%S"
    is_image = True
    separate_files = False
    all_bands = ["B02", "B03", "B04"]
    rgb_bands = ["B02", "B03", "B04"]
    
    def plot(self, sample):
        # Find the correct band index order
        rgb_indices = []
        for band in self.rgb_bands:
            rgb_indices.append(self.all_bands.index(band))

        # Reorder and rescale the image
        image = sample["image"][rgb_indices].permute(1, 2, 0)
        image = (image / 255).numpy() #torch.clamp(image / 10000, min=0, max=1).numpy()

        # Plot the image
        fig, ax = plt.subplots()
        ax.imshow(image)

        return fig
```

```python
def _crop_nodata(arr,nodata,img_shape):
    # find the indices of the non-zero elements in the image
    w_idx = np.where(np.any(arr != nodata, axis=(0,1)))[0]
    h_idx = np.where(np.any(arr != nodata, axis=(0, 2)))[0]

    # find the bounding box of the non-zero region
    top = h_idx[0]
    bottom = h_idx[-1] + 1
    left = w_idx[0]
    right = w_idx[-1] + 1

    # crop the image to remove the padding
    arr = arr[:, top:bottom, left:right]
    bottom = img_shape[1] - bottom
    right = img_shape[2] - right
    
    # top,bottom = bottom,top
    # left,right = right,left


    if sum([arr.shape[1], top,bottom]) < img_shape[1]:
        if  top > 0 and bottom == 0:
             top += img_shape[1] - sum([arr.shape[1], top]) 
        elif bottom > 0 and  top == 0:
             bottom += img_shape[1] - sum([arr.shape[1], bottom])
        elif  top > 0 and bottom > 0:
            top += img_shape[1] - (sum([arr.shape[1], top,bottom]) // 2)
            bottom += img_shape[1] - sum([arr.shape[1], top,bottom])
    if sum([arr.shape[2], left,right]) < img_shape[2]:
        if  left > 0 and right == 0:
             left += img_shape[2] - sum([arr.shape[2], left]) 
        elif right > 0 and  left == 0:
             right += img_shape[2] - sum([arr.shape[2], left,right])
        elif  left > 0 and right > 0:
            left += img_shape[2] - (sum([arr.shape[2], left]) // 2) 
            right += img_shape[2] - sum([arr.shape[2], left,right])
            
    if sum([top,bottom,left,right]) > 0:
        arr = np.pad(arr,[(0,0),(top,bottom),(left,right)],"reflect")
    return arr
```
